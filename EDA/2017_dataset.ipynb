# Preliminary Visualizations using 2017 census data

We have cheched the census data of 2017 and tried making few Preliminary Visualization.

Below are the few Visualizations we have made on 2017 data

# GDP per capita

```python
state_income2 = df2.groupby(['State']).mean(['IncomePerCap']).reset_index()[['State', 'IncomePerCap']]
state_income2.head()
```

```python
f = plt.figure()
f.set_figwidth(10)
f.set_figheight(5)
plt.bar(state_income2['State'], state_income2['IncomePerCap'], color = 'maroon')
plt.ylim(0,50000)
plt.xticks(fontsize = 7, rotation=90)
plt.xlabel("States")
plt.ylabel("GDP per capita")
plt.title("GDP per Capita by states")
```

# Total population

```python
df2['minority'] = round(((df2['Hispanic']+df2['Black']+df2['Native']+df2['Asian']+df2['Pacific'])*df2['TotalPop'])/100)
pop_diversity2 = df2.groupby(['State']).sum().sort_values('minority').reset_index()[['State', "TotalPop", 'minority']]
```

```python
plt.figure(figsize=(10,5))
plt.bar(pop_diversity2['State'], pop_diversity2['TotalPop'])
plt.ylim(0,45000000)
plt.xticks(fontsize = 7, rotation=90)
plt.xlabel("States")
plt.ylabel("Total population ( x10 million )")
plt.title("Total population by states")
```

# Total pop vs minority pop

```python
fig, ax1 = plt.subplots(figsize = (10, 5))
ax1.set_xlabel('State')
ax1.set_ylabel('Total pop', color = 'red')
ax1.plot(pop_diversity2['State'], pop_diversity2['TotalPop'], color = 'red')
ax1.tick_params(axis ='y', labelcolor = 'red')
ax1.tick_params(axis ='x', labelrotation = 90, labelsize = 7)

# Adding Twin Axes

ax2 = ax1.twinx()

ax2.set_ylabel('Minority pop', color = 'blue')
ax2.plot(pop_diversity2['State'],pop_diversity2['minority'] , color = 'blue')
ax2.tick_params(axis ='y', labelcolor = 'blue')
```

```python
fig, ax = plt.subplots(figsize = (10, 5))


ax2 = ax.twinx()

# creating a bar plot

ax.bar(pop_diversity2['State'], pop_diversity2["TotalPop"], color = 'red')
ax2.bar(pop_diversity2['State'], pop_diversity2["minority"], color = 'blue')

# Label axes

ax.set_xlabel('States')
ax.set_ylabel('Total State Population')
ax2.set_ylabel('Minority population of state')

# defining layout
plt.tight_layout()
fig.autofmt_xdate(rotation=90)

# show plot
plt.show()
```

# Mean commute by states

```python
commute = df2.groupby(['State']).mean('MeanCommute').reset_index()[['State', 'MeanCommute']]
commute.head()
```

```python
short_long = []
for x in commute['MeanCommute']:
    if x > 25 : 
        short_long.append('red')
    else:
        short_long.append('blue')

print(short_long)
```

```python
commute['commute_time'] = short_long
commute.head()
```

```python
plt.figure(figsize=(10,5))
plt.bar(commute['State'], commute['MeanCommute'], color = commute['commute_time'])
plt.ylim(0,35)
plt.xticks(fontsize = 7, rotation=90)
plt.xlabel("States")
plt.ylabel("Commute time ( in minutes )")
plt.title("Mean commute time vs States")
```
